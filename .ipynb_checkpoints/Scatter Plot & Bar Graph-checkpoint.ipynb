```python
import pandas as pd 
from matplotlib import pyplot as plt
from matplotlib import rcParams
```

```python
data=pd.read_csv("insurance.csv")
x=data['age']
y=data['buy']
```

```python
rcParams['figure.figsize'] = 10,8
plt.bar(y.unique(), y.value_counts(), color = ['green', 'orange'])
plt.xticks([0, 1])
plt.xlabel('Age,Buy')
plt.ylabel('Count')
plt.title('Count of each Target Class')
```

```python
rcParams['figure.figsize'] = 20,5
plt.scatter(x,y)
plt.xlabel("Age")
plt.ylabel("buy")
plt.xticks(x.unique())
plt.show()
```
